## Tools

Tools are an incredibly powerful feature provided by the frontier LLMs.

With tools, you can write a function, and have the LLM call that function as part of its response.

Sounds almost spooky.. we're giving it the power to run code on our machine?

Well, kinda.

```python
# Let's start by making a useful function

ticket_prices = {"london": "$799", "paris": "$899", "tokyo": "$1400", "berlin": "$499"}

def get_ticket_price(destination_city):
    print(f"Tool get_ticket_price called for {destination_city}")
    city = destination_city.lower()
    return ticket_prices.get(city, "Unknown")
```

```python
get_ticket_price("London")
```

```python
def book_ticket(destination_city, time_of_travel, passenger_name, flight_class='economy'):
    print(f"Your flight has been booked for {destination_city}, on {time_of_travel}");

    # Calculate ticket price based on flight class
    base_price = int(ticket_prices.get(destination_city.lower(), "0").replace("$", ""))
    if flight_class == "business":
        price = base_price * 1.5
    elif flight_class == "first":
        price = base_price * 2
    else:
        price = base_price
    
    # Create booking details
    booking_details = {
        "destination_city": destination_city,
        "time_of_travel": time_of_travel,
        "passenger_name": passenger_name,
        "flight_class": flight_class,
        "price": f"${price}",
    }
    return booking_details
```

```python
# There's a particular dictionary structure that's required to describe our function:

price_function = {
    "name": "get_ticket_price",
    "description": "Get the price of a return ticket to the destination city. Call this whenever you need to know the ticket price, for example when a customer asks 'How much is a ticket to this city'",
    "parameters": {
        "type": "object",
        "properties": {
            "destination_city": {
                "type": "string",
                "description": "The city that the customer wants to travel to",
            },
        },
        "required": ["destination_city"],
        "additionalProperties": False
    }
}
```

```python
booking_function = {
    "name": "book_ticket",
    "description": "Ticket booking function. Call this when a user wants to book a flight. For example, when a user wants to book a flight, run this method.",
    "parameters": {
        "type": "object",
        "properties": {
            "destination_city": {
                "type": "string",
                "description": "The city that the customer wants to travel to",
            },
            "time_of_travel": {
                "type": "string",
                "description": "The date of the flight (YYYY-MM-DD).",
            },
            "passenger_name": {
                "type": "string",
                "description": "Full name of the passenger.",
            },
             "flight_class": {
                "type": "string",
                "description": "Flight class (economy, business, first). Default is economy.",
                "enum": ["economy", "business", "first"],
            },
        },
        "required": ["destination_city", "time_of_travel", "passenger_name"],
    },
}
```

```python
# And this is included in a list of tools:

tools = [{"type": "function", "function": price_function},
        {"type": "function", "function": booking_function}]
```

## Getting OpenAI to use our Tool

There's some fiddly stuff to allow OpenAI "to call our tool"

What we actually do is give the LLM the opportunity to inform us that it wants us to run the tool.

Here's how the new chat function looks:

```python
def chat(message, history):
    messages = [{"role": "system", "content": system_message}] + history + [{"role": "user", "content": message}]
    response = openai.chat.completions.create(model=MODEL, messages=messages, tools=tools)

    if response.choices[0].finish_reason=="tool_calls":
        message = response.choices[0].message
        response, city = handle_tool_call(message)
        messages.append(message)
        messages.append(response)
        response = openai.chat.completions.create(model=MODEL, messages=messages)
    
    return response.choices[0].message.content
```

```python
# We have to write that function handle_tool_call:

# Function to handle tool calls
def handle_tool_call(message):
    tool_call = message.tool_calls[0]
    if tool_call.function.name == "get_ticket_price":
        arguments = json.loads(tool_call.function.arguments)
        destination_city = arguments.get("destination_city")
        price = get_ticket_price(destination_city)
        response = {
            "role": "tool",
            "content": json.dumps({"destination_city": destination_city, "price": price}),
            "tool_call_id": tool_call.id,
        }
        return response, destination_city

    elif tool_call.function.name == "book_ticket":
        arguments = json.loads(tool_call.function.arguments)
        
        destination_city = arguments.get("destination_city")
        time_of_travel = arguments.get("time_of_travel")
        passenger_name = arguments.get("passenger_name")
        flight_class = arguments.get("flight_class", "economy")
        
        booking_details = book_ticket(destination_city, time_of_travel, passenger_name, flight_class)
        response = {
            "role": "tool",
            "content": json.dumps(booking_details),
            "tool_call_id": tool_call.id,
        }
        return response, destination_city
```

Let's put this in a function called artist.

```python
# Some imports for handling images

import base64
from io import BytesIO
from PIL import Image
```

```python
def artist(city):
    image_response = openai.images.generate(
            model="dall-e-3",
            prompt=f"An image representing a vacation in {city}, showing tourist spots and everything unique about {city}, in a vibrant pop-art style",
            size="1024x1024",
            n=1,
            response_format="b64_json",
        )
    image_base64 = image_response.data[0].b64_json
    image_data = base64.b64decode(image_base64)
    return Image.open(BytesIO(image_data))
```

```python
import sounddevice as sd
import numpy as np
import scipy.io.wavfile as wav
import threading
import queue
import asyncio
import gradio as gr

# Global variables for recording control
is_recording = False
audio_queue = queue.Queue()

SAMPLE_RATE = 16000  # Sample rate of 16kHz is recommended for Whisper
DURATION = 10  # Duration of the recording in seconds
FILENAME = "recorded_audio.wav"  # Name of the output audio file

# Function to record audio in real-time
def record_audio_with_stream(sample_rate, filename):
    global is_recording
    print("Recording started...")
    
    def audio_callback(indata, frames, time, status):
        """Callback function to capture audio data in real-time."""
        if status:
            print("Audio recording status:", status)
        if is_recording:
            audio_queue.put(indata.copy())
    
    # Start recording
    with sd.InputStream(samplerate=sample_rate, channels=1, dtype='int16', callback=audio_callback):
        while is_recording:
            sd.sleep(100)  # Keep the stream alive while recording
    
    # Save the recorded audio to a WAV file
    audio_data = []
    while not audio_queue.empty():
        audio_data.append(audio_queue.get())
    audio_data = np.concatenate(audio_data, axis=0)
    wav.write(filename, sample_rate, audio_data)
    print(f"Recording stopped. Audio saved as {filename}")
```

```python
def transcribe_with_openai(filename):
    try:
        # Open the audio file in binary mode
        with open(filename, "rb") as audio_file:
            # Send the audio file to OpenAI's Whisper API for transcription
            transcription = openai.audio.transcriptions.create(
                model="whisper-1",  # Use the Whisper model
                file=audio_file
            )
            # Print the transcription result
            print("Transcription:")
            print(transcription.text)

            # Return the transcribed text
            return transcription.text
    except Exception as e:
        # Handle any errors that occur during transcription
        print("Error during transcription:", e)
        return "Transcription failed."
```

```python
# Gradio interface

# Gradio interface
async def toggle_recording():
    global is_recording
    if not is_recording:
        # Start recording
        is_recording = True
        threading.Thread(target=record_audio_with_stream, args=(SAMPLE_RATE, FILENAME)).start()
        return "Recording...", "Stop Recording"
    else:
        # Stop recording
        is_recording = False
        print("Recording stopped. Waiting for audio to be saved...")

        # Wait for the file to be written
        while not audio_queue.empty():
            await asyncio.sleep(0.1)  # Wait for the recording thread to finish

        print("Audio saved. Transcribing...")
        transcription = transcribe_with_openai(FILENAME)
        return transcription, "Start Recording"
```

# Our Agent Framework

The term 'Agentic AI' and Agentization is an umbrella term that refers to a number of techniques, such as:

1. Breaking a complex problem into smaller steps, with multiple LLMs carrying out specialized tasks
2. The ability for LLMs to use Tools to give them additional capabilities
3. The 'Agent Environment' which allows Agents to collaborate
4. An LLM can act as the Planner, dividing bigger tasks into smaller ones for the specialists
5. The concept of an Agent having autonomy / agency, beyond just responding to a prompt - such as Memory

We're showing 1 and 2 here, and to a lesser extent 3 and 5. In week 8 we will do the lot!

```python
# Define the chat function

def chat(history, translated_history):
    messages = [{"role": "system", "content": system_message}] + history
    response = openai.chat.completions.create(model=MODEL, messages=messages, tools=tools)
    image = None
    
    if response.choices[0].finish_reason == "tool_calls":
        message = response.choices[0].message
        response, city = handle_tool_call(message)
        messages.append(message)
        messages.append(response)
        image = artist(city)
        response = openai.chat.completions.create(model=MODEL, messages=messages)
        
    reply = response.choices[0].message.content
    history += [{"role": "assistant", "content": reply}]

    # Translate the reply
    translated_reply = translate(reply)
    
    # Convert string to list if needed, then append
    if isinstance(translated_history, str):
        if translated_history:  # If there's existing text
            translated_history = [translated_history]
        else:  # If empty string
            translated_history = []
    
    translated_history.append(translated_reply)
    
    # Comment out or delete the next line if you'd rather skip Audio for now..
    talker(reply)
    
    return history, image, "\n".join(translated_history)  # Return as joined string
```

```python
def do_entry(message, history, translated_history):
    # If the message is from voice transcription, use it directly
    if isinstance(message, str):
        history += [{"role": "user", "content": message}]
    return "", history, translated_history

# Gradio interface
with gr.Blocks() as ui:
    with gr.Row():
        chatbot = gr.Chatbot(height=500, type="messages")
        with gr.Column():
            image_output = gr.Image(height=300)
            translated_output = gr.Textbox(label="Translated Reply", lines=5, interactive=False)
    with gr.Row():
        entry = gr.Textbox(label="Chat with our AI Assistant:", placeholder="Type your message or use voice input...")
    with gr.Row():
        record_button = gr.Button("Start Recording")  # Single toggle button
        submit_button = gr.Button("Submit")  # New submit button
    with gr.Row():
        clear = gr.Button("Clear")

    # Define interactions
    # Submit button interaction
    submit_button.click(
        do_entry,
        inputs=[entry, chatbot, translated_output],
        outputs=[entry, chatbot, translated_output]
    ).then(
        chat,  # Replace with your chat function
        inputs=[chatbot, translated_output],
        outputs=[chatbot, image_output, translated_output]
    )

    # Entry submit interaction (when pressing Enter)
    entry.submit(
        do_entry,
        inputs=[entry, chatbot, translated_output],
        outputs=[entry, chatbot, translated_output]
    ).then(
        chat,  # Replace with your chat function
        inputs=[chatbot, translated_output],
        outputs=[chatbot, image_output, translated_output]
    )

    # Toggle recording button
    record_button.click(
        toggle_recording,
        outputs=[entry, record_button]  # Update entry and button text
    )

    # Clear button
    clear.click(
        lambda: ([], []),  # Reset both chatbot and translated history
        inputs=None,
        outputs=[chatbot, translated_output],
        queue=False
    )

ui.launch(inbrowser=True)
```
